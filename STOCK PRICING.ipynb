```python
import yfinance as yf 
import pandas as pd 
from datetime import datetime,timedelta
import os 
```

```python
# définir les indices et leur palge de temps 
tickers = ['SPY','BND','GLD','QQQ','VTI']
```

```python
# Fixer la date de fin à aujourd'hui 

end_date = datetime.today()
print(end_date)
```

```python
# Fixer la date de début il y'a 2 ans 

Start_date = end_date - timedelta(days=2*365)
print(Start_date)
```

SECTION 2 : TELECHARGER LES PRIX DE CLÔTURE

```python
close_df = pd.DataFrame()
```

```python
#télécharger les prix de clôtures pour chaque ticker 
for ticker in tickers:
    data = yf.download(ticker, start = Start_date,end = end_date)
    close_df[ticker] = data['Adj Close']
```

```python
print(close_df)
```
